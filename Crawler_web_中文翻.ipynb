```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import fake_useragent as ua
import time
import csv
import os
```

```python
# 创建 Chrome WebDriver 选项
chrome_options = Options()

# 设置请求头
chrome_options.add_argument("--user-agent=Your_User_Agent_String")

# 禁用 SSL 验证
chrome_options.add_argument("--ignore-certificate-errors")
chrome_options.add_argument("--charset=UTF-8")

# 创建 Chrome WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=chrome_options)
```

```python
for i in range(0, 1):
    csv_w = open('url/教育_關鍵.csv', "w", encoding='utf-8-sig', newline='')
    writer = csv.writer(csv_w)
    writer.writerow(["h2", "web"])
    url_list = ['https://www.thenewslens.com/category/education']
    
    
    for url in url_list:
        headers = {"User-Agent": ua.UserAgent().random}
        html = driver.get(url)
        
        #Change pages
        pagination = driver.find_element("xpath", '//ul[@class="pagination"]')     #找到換頁選單
        #init.
        current_page = 1
        last_page = 50
        title_temp = []
        url_temp = []
        
        while current_page <= last_page:    #限制要翻幾頁的loop
            for _ in range(5):      #進入新的page開始下拉，要算拉幾下會遇到pagination
                driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
                time.sleep(2)
                
            #把要爬的資料寫在這 
            soup = BeautifulSoup(driver.page_source, "html.parser")
            h2_tags = soup.find_all('h2', class_='title')
            # print(h2_tags)
            # a_tags = soup.find_all('a', class_ = 'GenericCard_link__EMXqX')
            for h2 in h2_tags:
                url = h2.find_next("a")["href"]
                title = h2.find_next("a")["title"]
                print(title,  url)
                title_temp.append(title)
                url_temp.append(url)
                
            # for a in a_tags:
            #     h3_tags_temp.append(a.text.strip())

            next_page = driver.find_element("xpath", '//a[@title="下一頁"]')
            next_page.click()
            print('done page'+ str(current_page))
            time.sleep(4)
            current_page = current_page + 1
            
        i=0
        for title in title_temp:
            writer.writerow([title_temp[i], url_temp[i]])
            i+=1
```
