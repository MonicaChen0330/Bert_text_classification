```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import fake_useragent as ua
import time
import csv
import os
```

```python
# 創建 Chrome WebDriver 選項
chrome_options = Options()

# 設置請求頭
chrome_options.add_argument("--user-agent=Your_User_Agent_String")

# 禁用SSL驗證
chrome_options.add_argument("--ignore-certificate-errors")
chrome_options.add_argument("--charset=UTF-8")

# 創建 Chrome WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=chrome_options)
```

```python
for i in range(0, 1):
    #csv_w = open(news_labels[i] + ".csv", "w", encoding="utf-8-sig")   # 合併網址
    csv_w = open('url/shopping_yahoo.csv', "w", encoding='utf-8-sig', newline='')    # 開啟csv，若不存在該csv檔會自動建立，w覆蓋寫入，a附加寫入
    writer = csv.writer(csv_w)
    writer.writerow(["h2", "web"])  #寫標題行，若用a寫入則不須，否則戶重複寫在中間
    url_list = ['https://www.yahoo.com/lifestyle/tagged/shopping/']
    for url in url_list:
        headers = {"User-Agent": ua.UserAgent().random}     # 生成新的header
        html = driver.get(url)      # 抓取網頁程式
        for _ in range(500):
            driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)  # 模擬按下 Page Down 鍵
            time.sleep(2)    # 等待一段時間，讓新內容載入
        soup = BeautifulSoup(driver.page_source, "html.parser")     # 整理網頁程式
            
        h3_temp = []
        h3_tags = soup.find_all("h3")   # 找到所有h3
        for h3_tag in h3_tags:  # 遍歷每個 h3 標籤
            # 找到 h3 標籤下的相關網址
            # 這裡假設相關的網址是 h3 標籤後面的<a>標籤
            related_url = h3_tag.find_next("a")["href"] if h3_tag.find_next("a") else None
            # 打印相關網址 與標題
            if related_url:
                print(h3_tag.text.strip())
                print(related_url)
                h3_temp.append(related_url)
                writer.writerow(
                    [h3_tag.text.strip(), (related_url).replace("\n", "")]
                )
        print(len(h3_temp))
```
